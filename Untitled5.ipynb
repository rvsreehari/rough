```python
import os
import re
import pdfplumber

root_folder = "path/to/folder"
pqcheck = ["eligibility criteria", "financial statement", "year", "certification", "turnover", "litigation history", "general", "pq", "oem", "it", "qualification", "jv/ consortium"]
condcheck = ["qualification", "criteria", "requirement", "capacity", "statement", "history", "factor", "ipr"]
condcheck1 = ["^\d+\.", "^\d+$", "^([A-Z]+\s)*[A-Z]+$"]  # serial number, number, all capital letters

pq_dict = {
    "JV/ Consortium": ["joint venture", "jvc", "joint-venture", "affiliate", "conflict of interest"],
    "OEM": ["oem's", "osd"],
    "Qualification": ["qualifying", "qualification"],
    "Turnover": ["annual turnover", "last <no.> year turnover", "past <no.> turnover", "current year turnover", "fy turnover"],
    "Certification": ["experience certificates", "experience criteria", "commercial compliance certificate", "compliance certificate", "performance certificate", "certificate of incorporation", "experience", "relevant certificate", "relevant certification", "self-attested", "self attested", "performance certificate", "quality certification", "quality certificate"],
    "Financial": ["financial qualification", "financial requirement", "financial capacity", "financial criteria", "financial statement", "financial statements"],
    "Year": ["year", "fy", "financial year"],
    "IT": ["income tax return", "it return"],
    "eligibility criteria": ["eligibility criteria", "minimum eligibility criteria", "pre-qualifi cation", "pre qualification", "pre-qualification", "pq"],
    "General": ["enhancement factor", "ipr"],
    "Litigation": ["litigation history", "past litigation", "blacklist", "terminate", "hold", "termination"]
}

for subdir, dirs, files in os.walk(root_folder):
    for file in files:
        filepath = os.path.join(subdir, file)
        if filepath.endswith(".pdf"):
            try:
                with pdfplumber.open(filepath) as pdf_file:
                    for page_num, page in enumerate(pdf_file.pages):
                        if not page.extract_text():
                            continue  # Skip this page if it has no text
                        for pq_keyword in pqcheck:
                            if pq_keyword in page.extract_text().lower():
                                for lt_obj in page.layout:
                                    text = lt_obj.get_text().lower()
                                    if any([re.search(cond, text, re.IGNORECASE) for cond in condcheck+condcheck1]):
                                        bold = lt_obj.fontname.endswith("Bold")
                                        underline = any(underline for _, _, underline in lt_obj._text_runs)
                                        serial_num = re.match("^\d+\.", text)
                                        capital_letters = re.match("^([A-Z]+\s)*[A-Z]+$", text)
                                        if bold or underline or serial_num or capital_letters:
                                            search_terms = set([term.lower() for term in pq_dict[pq_keyword]])
                                            matched_terms = [term for term in search_terms if term in text]
                                            if matched_terms:
                                                print(f"Matched keywords: {pq_keyword}, Page number: {page_num+1}")
            except:
                continue
```

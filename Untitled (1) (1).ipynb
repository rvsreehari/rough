```python
from transformers import pipeline

model_name = "bert-base-uncased"


classifier = pipeline("zero-shot-classification", model=model_name)
```

```python
# Classify the text into the predefined categories
classifier = pipeline("zero-shot-classification")
results = classifier(text, categories)

# Extract the text for each category label
category_texts = {}
for label in categories:
    if label in results["labels"]:
        category_text = ""
        for i, score in enumerate(results["scores"]):
            if results["labels"][i] == label:
                category_text += results["sequence"] + " "
        category_texts[label] = category_text.strip()

# Print the extracted text for each category
for label, category_text in category_texts.items():
    print(f"{label}: {category_text}")
```

```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments

# Load the pre-trained GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python

# Load your text data
with open(r"C:\Users\rvsre\Desktop\hi.txt", encoding='utf-8') as f:
    text = f.read()

```

```python
# Tokenize the text data
tokenized_text = tokenizer.encode(text)

```

```python
# Convert the tokenized text to a PyTorch dataset
max_seq_length = 6,400
dataset = TextDataset(tokenized_text, tokenizer,block_size=max_seq_length)

```

```python
# Set up the data collator
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

```

```python
# Set up the training arguments
training_args = TrainingArguments(
    output_dir='./results',
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=8,
    save_steps=1000,
    save_total_limit=2
)

```

```python
# Set up the trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    data_collator=data_collator
)

```

```python
# Fine-tune the model on your text data
trainer.train()

```

```python
# Save the fine-tuned model
model.save_pretrained('your_model_dir')
```

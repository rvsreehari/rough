```python
import fitz
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
import re

# download required resources
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("wordnet")

# define pre-defined keywords
keywords = ['pre-qualification', 'eligibility criteria', 'pre – qualification', 'pre-qualification', 'pq criteria', 
           'prequalification', 'pre -qualification', 'pre -qualification detail', 'conditions of eligibility of bidder', 
           'pre-qualification performa', 'evaluation and qualification criteria', 'qualifying criteria', 'eligible bidders', 
           'qualification criteria','minimum eligibility criteria','experience criteria','financial criteria',
            'enhancement factor','litigation history']


# open the PDF file
doc = fitz.open(r"C:/Users/rv.sreehari/Desktop/Structured/defproc_19012023_010426881425.pdf")

text = ""

# # iterate over the pages and extract text
# for page in doc:
#     text += page.get_text()

# # close the PDF file
# doc.close()


text = ""

# iterate over the pages and extract text
for page in doc:
    text += page.get_text()

# # close the PDF file
# doc.close()

# load stop words
stop_words = set(stopwords.words("english"))

# initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# remove punctuation and convert to lowercase
text = text.translate(str.maketrans("", "", string.punctuation)).lower()

# tokenize text into words
words = word_tokenize(text)

# remove stop words and lemmatize words
words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

# join words into a string
processed_text = " ".join(words)

# initialize empty list to store relevant context
relevant_context = []

# iterate over sentences in the text data
for sentence in text.split("."):
    # check if any of the pre-defined keywords are present in the sentence
    if any(keyword in sentence for keyword in keywords):
        # append the sentence to the relevant context list
        relevant_context.append(sentence)

# print the relevant context in a list format
for i, sentence in enumerate(relevant_context):
    print(f"{i+1}. {sentence}")
```
